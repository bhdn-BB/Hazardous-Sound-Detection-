```python
import pandas as pd
import dotenv
import os
import re
```

```python
dotenv.load_dotenv()

BASE_PATH = os.getenv('BASE_PATH')

TRAIN_DATA_PATH = os.path.join(BASE_PATH, 'data', 'train')
TEST_DATA_PATH = os.path.join(BASE_PATH, 'data', 'test')

TRAIN_SAMPLES_NAME = os.listdir(TRAIN_DATA_PATH)
TEST_SAMPLES_NAME = os.listdir(TEST_DATA_PATH)

TRAIN_PATHS = [os.path.join(TRAIN_DATA_PATH, name) for name in TRAIN_SAMPLES_NAME]
TEST_PATHS = [os.path.join(TEST_DATA_PATH, name) for name in TEST_SAMPLES_NAME]

LABELS = ['siren', 'gunshot', 'explosion', 'casual']
label2id = {label: i for i, label in enumerate(LABELS)}
label2id
```

```python
TRAIN_PATHS
```

```python
pattern = re.compile(r'\\([a-zA-Z0-9_]+)_\d+\.wav$')

data = []
for path in TRAIN_PATHS:
    match = pattern.search(path)
    if match:
        cls_name = match.group(1)
        if cls_name in label2id:
            target = label2id[cls_name]
            data.append((path, target))
df = pd.DataFrame(data, columns=["path", "target"])
df
```

```python
output_path = os.path.join(BASE_PATH, 'data', 'metadata.csv')
df.to_csv(output_path, index=False)
```
